```python
import turtle         #터틀 그래픽 모듈을 불러온다
import random         #랜덤 모듈을 불러옵니다
t = turtle.Turtle()   #거북이를 만듭니다
t.shape("turtle")     #커서의 모양을 거북이로 설정한다

for i in range(30):   #30번 반복한다
    length = random.randint(1,100) #1부터 100사이의 난수를 생성하여 length변수에 저장한다
    t.forward(length)              #length만큼 전진한다
    angle = random.randint(1,4)    #1부터 4사이의 난수를 생성하여 angle변수에 저장한다
    t.right(90*angle)              #90*angle만큼 오른쪽으로 회전한다
```

```python
import turtle #터틀 그래픽 모듈을 불러온다
import random #랜덤 모듈을 불러온다

#색상은 리스트에 저장했다가 하나씩 꺼내서 변경하도록 하자
colors = ["red","purple","blue","green","yellow","orange"]
t = turtle.Turtle()     #거북이를 만든다
turtle.bgcolor("black") #배경색을 검정으로 한다
t.speed(0)              #거북이의 속도는 0으로 설정하면 최대가 된다
t.width(3)              #거북이의 두께를 3으로 한다
length = 10             #초기선의 길이는 10으로 한다

while length < 500:              #while은 반복문이다. 선의길이가 500보다 작으면 반복한다
  t.forward(length)              #length만큼 전진한다
  t.pencolor(colors[length%6])   #선의 색상을 변경한다
  angle = random.randint(1,100)  #1부터 100사이의 난수를 생성하여 angle 변수에 저장한다
  t.right(angle)                 #angle만큼 우회전한다
  length += 5                    #선의 길이를 5만큼 증가한다
```
